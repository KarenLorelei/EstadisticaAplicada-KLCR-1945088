```python
import matplotlib.pyplot as plt
plt.scatter(df1['walmart'], df1['soriana'], c=df1['aurrera'], cmap = 'coolwarm')
plt.colorbar().set_label('aurrera')
plt.xlabel('walmart')
plt.ylabel('soriana')
plt.show
```
