```python
#Renombrar columnas
# dataframe.rename(columns={'Nombre actual': 'Nombre Nuevo'})
calificaciones.rename(columns={'Calficación Final': 'Cal Final'}, inplace = True)
```

```python
calificaciones.head()
```

```python
calificaciones.rename(columns={'Id': 'id'}, inplace = True)
calificaciones.rename(columns={'Nombre': 'nombre'}, inplace = True)
calificaciones.rename(columns={'Apellido': 'apellido'}, inplace = True)
calificaciones.rename(columns={'Cal Final': 'cal final'}, inplace = True)
```

```python
calificaciones.head()
```

```python
calificaciones['cal final'].mean()
```

```python
# funcion SI
# vamos a añadir una nueva columna que nos diga si el alumno está aprobado o no
# dataframe['nombre de nueva columna'] = np.where(dataframe['columna de objeto a tomar'] condición , 'Valor si se cumple', 'Valor si no se cumple')
calificaciones['situación']=np.where(calificaciones['cal final']>= 70, 'Aprobado', 'Reprobado')
```

```python
calificaciones.head()
```

```python
# Cuenta los grupos de la columna situación, y cuantos hay, y solo muestra la cal final
calificaciones.groupby('situación').count()['cal final']
```

```python
#Cuenta los grupos de la columna situación, y cuantos hay en cada uno
calificaciones.value_counts('situación')
```

```python
# Saca el promedio de cada grupo
calificaciones.groupby('situación').mean()['cal final']
```

```python
#Hace un objeto con los alumnos mejores
excelentes = calificaciones[calificaciones['cal final']>= 90]
```

```python
excelentes
```

```python
#Ordenar de menor a mayor
calificaciones.sort_values('cal final',ascending=True)
```

```python
#Ordenar de mayor a menor
calificaciones.sort_values('cal final',ascending=False)
```

```python
#Ordenar por Apellido de A a Z
calificaciones.sort_values('apellido',ascending=True)
```

```python
#Ordenar por Apellido de A a Z y luego por nombre de A a Z
calificaciones.sort_values(['apellido','nombre'],ascending=[True,True])
```

```python
# Guardar orden anterior
calificaciones = calificaciones.sort_values(['apellido','nombre'],ascending=[True,True])
```
