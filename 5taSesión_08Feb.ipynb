```python
##Creamos DataFrame 
mun_nl = pd.DataFrame({'municipio':['Monterrey', 'Guadalupe', 'Apodaca', 'San Nicolás', 'Santiago'],
                      'nacimientos':[500, 630, 768, 400, 250]})
```

```python
mun_nl
```

```python
## Hacemos Gráfica de barras ordenada de mayor a menor

#Para ordenar datos
mun_nl = mun_nl.sort_values('nacimientos', ascending=False)
#Para Graficar
plt.bar(mun_nl['municipio'],
        mun_nl['nacimientos']);
#Para enseñar la gráfica
plt.show()
```

```python
#Gráfica Simple con colores
plt.bar(mun_nl['municipio'],       ##eje x
       mun_nl['nacimientos'],      #eje y
       color = ['red', 'orange', 'yellow', 'green','blue']);  ##color de cada barra

#Agregar título a la gráfica
plt.title('Nacimientos por municipio');

#Agregar nombre al eje x
plt.xlabel('Municipios de NL');

#Agregar nombre al eje y
plt.ylabel('Cantidad de nacimientos');

#Cambiar los saltos escala del eje y
plt.yticks([0,200,400,600,800]);

#Mostrar gráfica
plt.show()
```

```python
# fig es el nombre del lienzo
##fig,(nombresubgraficas)=plt.subplots(filas,columnas en que se ordenan las gráficas) 
fig, (panel1, panel2)=plt.subplots(2,1) 
fig.tight_layout()  ##ajusta las escalas de las gráficas para que tengan las mismas proporciones/tamaño

# panel1 (subgráfica 1)
x1 = x  ## eje x de gráfica 1
y1 = y  ## eje y de gráfica 1
panel1.plot(x,y,label='nombre de línea')
panel1.title.set_text('Gráfica 1')

# panel2 (subgráfica 2)
x2 = mun_nl['municipio']    ## eje x de gráfica 2
y2 = mun_nl['nacimientos']  ## eje y de gráfica 2
panel2.bar(x2,y2)
panel2.title.set_text('Gráfica 2')

plt.show()
```

```python
###Ahora con 4 Gráficas
# fig es el nombre del lienzo
##fig,(nombresubgraficas)=plt.subplots(filas,columnas en que se ordenan las gráficas) 
fig, (panel1, panel2, panel3, panel4)=plt.subplots(4,1) 
fig.tight_layout()  ##ajusta las escalas de las gráficas para que tengan las mismas proporciones/tamaño

# panel1 (subgráfica 1)
x1 = x  ## eje x de gráfica 1
y1 = y  ## eje y de gráfica 1
panel1.plot(x,y,label='nombre de línea',color = 'red')
panel1.title.set_text('Gráfica 1')

#panel2 (sungrafica 2)
x2 = x  ## eje x de gráfica 2
y2 = y  ## eje y de gráfica 2
panel2.scatter(x,y,label='nombre de línea',color = ['red', 'orange', 'yellow', 'green','blue'])
panel2.title.set_text('Gráfica 2')

# panel3 (subgráfica 3)
mun_nl = mun_nl.sort_values('nacimientos', ascending=False)
x3 = mun_nl['municipio']    ## eje x de gráfica 3
y3 = mun_nl['nacimientos']  ## eje y de gráfica 3
panel3.bar(x3,y3,color = ['red', 'orange', 'yellow', 'green','blue'])
panel3.title.set_text('Gráfica 3')

# panel4 (subgráfica 4)
mun_nl = mun_nl.sort_values('nacimientos', ascending=True)
x4 = mun_nl['municipio']    ## eje x de gráfica 4
y4 = mun_nl['nacimientos']  ## eje y de gráfica 4
panel4.barh(x4,y4,color = ['pink', 'skyblue', 'yellow', 'green','blue'])
panel4.title.set_text('Gráfica 4')

plt.show()
```
