```python
#las de siempre
import numpy as numpyp
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#liberias especializadas
import yfinance as yf
```

```python
# datos
ticker = 'AMZN.MX'
amazon = yf.download(ticker, 
                     start='2019-01-01', 
                     end='2023-03-17',
                     progress=False)
```

```python
amazon.head()
```

```python
amazon = amazon[['Adj Close']]
```

```python
amazon = amazon.rename(columns={'Adj Close': 'precio_cierre'})
```

```python
amazon['precio_cierre'].plot()
plt.title('Precio de la acción de Amazon \n 2019-2023')
plt.ylabel('pesos MXN')
plt.xlabel('fecha')
plt.show()
```

```python
# media movil a 10 dias
amazon['MA-10'] = amazon['precio_cierre'].rolling(window=10).mean()
```

```python
# media movil a 60 dias
amazon['MA-60'] = amazon['precio_cierre'].rolling(window=60).mean()
```

```python
amazon[['MA-10','MA-60']].plot()
plt.title('Señal de Largo plazo vs Corto plazo')
plt.ylabel('pesos MXN')
plt.xlabel('fecha')
plt.show()
```
