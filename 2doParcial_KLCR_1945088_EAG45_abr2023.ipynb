## 1. El banco sospecha que entre mas pasa el tiempo más díficil es que puedan pagar los clientes que se les concedió un préstamo. 

### a) Calcula el promedio del monto del préstamo de los que pagaron y el promedio del monto de los que no han pagado.

### b) Ajusta un modelo weibull y un modelo log-normal y graficas las funciones de riesgo respectivas. 


```python
# a)
pag = datos.query("full_repaid==1")
npag = datos.query('full_repaid==0')

pmpag = pag['amount'].mean()
pmnpag = npag['amount'].mean()

print('El promedio del monto del préstamo de los que pagaron es de', pmpag.round(2))
print('Y el promedio del monto del préstamo de los que no han pagado es de', pmnpag.round(2))
```

```python
!pip install lifelines
```

```python
# b)
from lifelines import WeibullFitter
weib = WeibullFitter().fit(datos['duration'], datos['full_repaid'])
weib.plot_hazard(label='Estimador Weibull', color= 'palevioletred')

from lifelines import LogNormalFitter
logn = LogNormalFitter().fit(datos['duration'], datos['full_repaid'])
logn.plot_hazard(label='Estimador Log-Normal', color= 'paleturquoise')

plt.title('Funciónes de riesgo de pagar la deuda Weibull vs Log-Normal')
plt.ylabel('$h(t)$')
plt.xlabel('meses')
plt.show()
```

```python
weib.plot_hazard(label = 'Estimador Weibull', color = 'mediumorchid');
plt.title('Función de riesgo de pagar la deuda')
plt.ylabel('$h(t)$')
plt.xlabel('meses')
plt.show()
weib.hazard_
```

```python
logn.plot_hazard(label = 'Estimador Log-Normal', color = 'palegreen');
plt.title('Función de riesgo de pagar la deuda')
plt.ylabel('$h(t)$')
plt.xlabel('meses')
plt.show()
logn.hazard_
```

### c) comenta que significa la grafica de la función de riesgo de la weibull respecto a si los clientes tardan en pagar

La grafica significa que conforme avanza el tiempo, el riesgo de pagar la deuda aumenta, lo que significa que entre mas tiempo pase, mas probable es que la deuda sea liquidada.

### d) comenta que significa la grafica de la función de riesgo de la log-normal respecto a si los clientes tardan en pagar

La grafica de log-normal me dice que entre los 20 y 30 meses es mas probable que se pague la deuda, y a partir de ese tiempo, cada vez es menos probable que suceda, o sea que entre mas tiempo pase, es mas dificil que las personas paguen su deuda.

## 2. Estima la función de supervivencia por el método de Kaplan-Meier para el tiempo que tardan los clientes en pagar y muestra su gráfica correspondiente

```python
from lifelines import KaplanMeierFitter
km = KaplanMeierFitter().fit(datos['duration'], datos['full_repaid'])
km.plot_survival_function(label = 'Tiempo que tardan los clientes en pagar', color = 'lightskyblue')
plt.title('Función de supervivencia de pagar la deuda, \n Intervalos al 95%')
plt.ylabel('$S(t)$')
plt.xlabel('meses')
plt.show()
km.survival_function_
```

```python
km.event_table
```

## 3. Coloca la gráfica de kaplan Meier sobre la grafica de supervivencia log-normal 

```python
logn.plot_survival_function(label = 'Log-Normal', color = 'palegreen');
km.plot_survival_function(label = 'Kaplan Meier', color = 'lightskyblue')
plt.title('Función de pagar la deuda')
plt.xlabel('meses')
plt.ylabel('S(t)')
plt.show()
```

## 4. Comenta si las graficas se parecen o no, y en que se diferencian

Las graficas si se parecen, pues a largo plazo van teniendo el mismo comportamiento. ## 5. El banco considera que los que piden préstado una gran cantidad de dinero en realidad nunca piensa pagar

### a) calcula la mediana de la cantidad que pidieron prestado (amount of money borrowed), asigna un 1 a quienes se encuentran igual o arriba de la mediana (deben mucho) y 0 a los que se encuentran debajo de la mediana (deben poco)

### b) Elabora y grafica juntas las funciones de supervivencia por el método de Kaplan-Meier para los que deben mucho y para los que deben poco.

### c)Interpreta las graficas de supervivencia, ¿qué significan? ¿quién paga más rápido? ¿quién tarda en pagar más?  

### d)Elabora un log-rank test para comprobar si las funciones de supervivencia de los que deben mucho y los que deben poco son distintas, e interpreta el resultado de la prueba

### e) ¿el banco tenía razón? Explique

```python
# a)
medat = datos['amount'].median()
datos['mayoimed']=np.where(datos['amount']>= medat, 1,0)
```

```python
datos.head()
```

```python
# b)
dm = datos.query("mayoimed==1")
dp = datos.query('mayoimed==0')

from lifelines import KaplanMeierFitter
kmm = KaplanMeierFitter().fit(dm['duration'], dm['full_repaid'])
kmm.plot_survival_function(label = 'Clientes que deben mucho', color = 'mediumorchid')

kmp = KaplanMeierFitter().fit(dp['duration'], dp['full_repaid'])
kmp.plot_survival_function(label = 'Clientes que deben poco', color = 'palevioletred')

plt.title('Función de supervivencia de pagar la deuda, \n Intervalos al 95%')
plt.ylabel('$S(t)$')
plt.xlabel('meses')
plt.show()
```

```python
kmm.survival_function_
```

```python
kmp.survival_function_
```

 c)

La grafica nos dice que los clientes que deben mucho demoran mas en pagar la deuda,ya que la funcion esta mas a la derecha; y es algo que se puede suponer desde el inicio, pues por logica es mas dificil pagar una deuda grande a una chica. Ademas de que siempre es mas probable que los de deben mucho no pagen su deuda a comparacion de los de deben poco; ya que la funcion de los que deben mucho siempre esta por arriba de los que deben poco.
 Por lo que los clientes que deben poco pagan mas rapido y los que deben mucho tardan mas en pagar.

```python
# d)
from lifelines.statistics import logrank_test
lrt = logrank_test(dm['duration'],dp['duration'],
                       dm['full_repaid'],dp['full_repaid'])
lrt.print_summary()
```

$H_0:$ Las funciones de supervivencia de los que deben mucho y los que deben poco son iguales.

 $H_a:$ Las funciones de supervivencia de los que deben mucho y los que deben poco son distintas.

Si p-value < $\alpha$, rechazamos $H_0$

$p$<0.0005 < $\alpha$=0.01

Por lo tanto, con un 99% de confianza rechazamos $H_0$, lo que significa que las funciones de supervivencia de los que deben mucho y los que deben poco son diferentes, o sea que su distribucion es distinta.

e)

El banco tiene razón en que los que piden grandes cantidades prestadas no suelen pagar tan rápido a comparación de los que deben sumas pequeñas; ya que vimos que la función de supervivencia de pagar las deudas de las personas que deben mucho siempre está por arriba y más a la derecha que la función de los clientes que deben poco.
